## Bag of Words ##

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
annak = ['Happy families are all alike;', 'every unhappy family is unhappy in its own way']
vect = CountVectorizer(max_features=1000)
vect.fit(annak)
# sparse matrix
X = vect.transform(annak)
# transform to an array
my_array = X.toarray()
print(my_array)
#transofrm back to a dataframe, assign column values
X_df = pd.DataFrame(my_array, columns=vect.get_feature_names())
```

## guess the languange ##

```python
from langdetect import detect_langs
foreign = '*brb cek kang nasi gorengnya bawa HT nggak?'
detect_langs(foreign)
```

```python
ih = pd.read_excel('indihome_dikit.xlsx')
ih_tweet = pd.DataFrame()
ih_tweet['tweet'] = ih.tweet
```

```python
ih_tweet.head()
```

```python
languanges = []
for row in range(len(ih_tweet.tweet)):
    languanges.append(detect_langs(ih_tweet.iloc[row, 0]))
```

```python
languanges = [str(languanges[0]).split(':')[0][1:] for lang in languanges]
```

```python
ih_tweet['languange'] = languanges
```

```python
ih_tweet.head()
```

## stop words ##

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
# define the stopwords list
my_stopwords = set(STOPWORDS)
my_stopwords.update(['sih', 'kok'])
```

```python
# generate and show the word cloud
corpus = ih_tweet.tweet.tolist()
corpus_str = ' '.join(map(str, corpus))
my_cloud = WordCloud(background_color='white', stopwords=my_stopwords).generate(corpus_str)
plt.imshow(my_cloud, interpolation='bilinear')
```

## stemming & lemmatization ##

```python
# lemma needs PoS
from nltk.stem import PorterStemmer
porter = PorterStemmer()
porter.stem('wonderful')
```

```python
from nltk.stem.snowball import SnowballStemmer

dutch_stemmer = SnowballStemmer("dutch")
dutch_stemmer.stem('beginnen')
```

```python
# how to stem
tokens = word_tokenize('today is wonderful day')
stemmed_tokens = [porter.stem(token) for token in tokens]
stemmed_tokens
```

```python
# lemmatization
from nltk.stem import WordNetLemmatizer
WNlemmatizer = WordNetLemmatizer()
WNlemmatizer.lemmatize('wonderful', pos='a')

# PoS mungkin bisa dipakai pada proses pre-process (dengan NER)
```
